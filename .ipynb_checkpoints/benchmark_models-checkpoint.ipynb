### the according plots, first for 2.1 MeV data

```python
fig, axs = plt.subplots(1,5,figsize=(14,6))
# Number of groups
num_groups = 1  # Since you only want to plot the first entry of each list

# X-axis positions for the bars
x = np.arange(num_groups)  # Single group at position 0
width = 0.2  # Width of the bars
fig.tight_layout()
for i in range(5):
    axs[i].bar(x - 2*width, r3b_2_1[i], width, label='r3b')
    axs[i].bar(x - width, agglo_2_1[i], width, label='agglo')
    axs[i].bar(x , edge_only_time_2_1[i], width, label='edge')
    axs[i].bar(x + width, agg_ed1_2_1[i], width, label='agglo+egde ver1')
    axs[i].bar(x + 2*width, agg_ed2_2_1[i], width, label='agglo+egde ver2')
    axs[i].legend()
for i in range (2,5):
    axs[i].set_ylim(0, 0.3)
axs[0].set_title("well_reco")
axs[1].set_title("true_positive")
axs[2].set_title("false_positive")
axs[3].set_title("false_negative")
axs[4].set_title("false mixed")
fig.suptitle("2.1MeV")
fig.tight_layout()

```
